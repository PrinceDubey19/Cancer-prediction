```python
X = df.drop(columns=['diagnosis'], axis = 1)
Y = df['diagnosis']
```

```python
df.columns
```

```python
# data modeling
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
# Algorithms Implementation
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
a=accuracy_score(ytest,ypred)
cm=confusion_matrix(ytest,ypred)
cr=classification_report(ytest,ypred)
print(a)
print(cm)
print(cr)
```

```python
 # by applying decision tree classifier we get the accuracy of 94%.
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(max_depth=15,n_estimators=70)
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
a=accuracy_score(ytest,ypred)
cm=confusion_matrix(ytest,ypred)
cr=classification_report(ytest,ypred)
print(a)
print(cm)
print(cr)
```

```python
# by applying random forest classifier we get the accuracy of 97%.
```

```python
'''
By applying two machine learning algorithm like decision tree and random forest,
we got decision tree classifier accuracy as 94%
random forest classifier as 97%.

from this we conclude that random forest classifier is the best algorithm for our model 
with the accuracy of 97%.
'''
```
